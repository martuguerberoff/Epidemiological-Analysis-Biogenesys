```python
#Ordenar datos por lugar y fecha para asegurar la secuencia correcta
df = df.sort_values(by=['location_key', 'date'])

# Recalcular el acumulado de Confirmados
# Agrupar por lugar y aplicar suma acumulativa a los nuevos casos
df['cumulative_confirmed'] = df.groupby('location_key')['new_confirmed'].cumsum()

#Recalcular el acumulado de Recuperados
df['cumulative_recovered'] = df.groupby('location_key')['new_recovered'].cumsum()

# Verificación rápida: comparar el último día para ver los nuevos totales
cols_verificacion = ['date', 'location_key', "new_confirmed", 'cumulative_confirmed']
print(df[cols_verificacion].tail())
```

        4.2. Comprobación de posibles anomalías

```python
print("--- INICIO DE AUDITORÍA DE ANOMALÍAS ---\n")

# 1. VERIFICACIÓN DE LÓGICA EPIDEMIOLÓGICA
# Buscar registros donde las muertes superan a los confirmados (imposible)
mask_logic_death = df['cumulative_deceased'] > df['cumulative_confirmed']
errors_logic_death = df[mask_logic_death]

# Buscar registros donde los recuperados superan a los confirmados
mask_logic_rec = df['cumulative_recovered'] > df['cumulative_confirmed']
errors_logic_rec = df[mask_logic_rec]

print(f"[Lógica] Muertes Acumuladas > Confirmados Acumulados: {len(errors_logic_death)} casos detectados.")
if len(errors_logic_death) > 0:
    print(errors_logic_death[['date', 'location_key', 'cumulative_confirmed', 'cumulative_deceased']].head())

print(f"[Lógica] Recuperados Acumulados > Confirmados Acumulados: {len(errors_logic_rec)} casos detectados.")


# 2. VERIFICACIÓN DE DUPLICADOS
# Verificar si existe más de una fila para la misma clave y fecha
duplicates = df.duplicated(subset=['location_key', 'date'], keep=False)
num_duplicates = duplicates.sum()

print(f"\n[Estructura] Filas duplicadas (mismo lugar y fecha): {num_duplicates}")


# 3. VERIFICACIÓN DE DATOS DEMOGRÁFICOS
# Buscar lugares con población 0 o nula (crítico para cálculos per cápita)
mask_pop_zero = (df['population'] == 0) | (df['population'].isna())
num_pop_issues = mask_pop_zero.sum()

print(f"\n[Demografía] Población 0 o vacía: {num_pop_issues} registros.")


# 4. DETECCIÓN DE PICOS EXTREMOS (OUTLIERS)
# Un método simple: Casos nuevos mayores que la población total (Error grave)
# Se usa 'population' como límite lógico superior
mask_extreme = (df['new_confirmed'] > df['population']) & (df['population'] > 0)
errors_extreme = df[mask_extreme]

print(f"\n[Estadística] Nuevos casos > Población total: {len(errors_extreme)} casos detectados.")
if len(errors_extreme) > 0:
    print(errors_extreme[['date', 'location_key', 'new_confirmed', 'population']].head())

print("\n--- FIN DE AUDITORÍA ---")
```

        4.2.1 Corrección de dichas anomalías

```python
# Corrección 1: Si hay más muertos que confirmados, se asume que los confirmados estaban subreportados
# Igualar confirmados a muertos en esos casos
mask_muertes = df['cumulative_deceased'] > df['cumulative_confirmed']
df.loc[mask_muertes, 'cumulative_confirmed'] = df.loc[mask_muertes, 'cumulative_deceased']

# Corrección 2: Si hay más recuperados que confirmados, se actualiza confirmados
# Igualar confirmados a recuperados en esos casos
mask_recuperados = df['cumulative_recovered'] > df['cumulative_confirmed']
df.loc[mask_recuperados, 'cumulative_confirmed'] = df.loc[mask_recuperados, 'cumulative_recovered']

# Verificación final de limpieza
# Contar los errores para asegurar que den 0
errores_muertes = (df['cumulative_deceased'] > df['cumulative_confirmed']).sum()
errores_recuperados = (df['cumulative_recovered'] > df['cumulative_confirmed']).sum()

print(f"Errores remanentes (Muertes > Confirmados): {errores_muertes}")
print(f"Errores remanentes (Recuperados > Confirmados): {errores_recuperados}")
```

```python
# Recalcular new_confirmed basándonos en los acumulados corregidos
# Ordenar primero para asegurar la resta correcta
df = df.sort_values(by=['location_key', 'date'])

# Calcular la diferencia del acumulado corregido
# Usamos fillna(0) para el primer día de cada país
df['new_confirmed'] = df.groupby('location_key')['cumulative_confirmed'].diff().fillna(0).astype('Int64')

# Corregir posibles negativos generados por el recálculo (si hubo ajustes de datos históricos)
df.loc[df['new_confirmed'] < 0, 'new_confirmed'] = 0

print("Columna 'new_confirmed' recalculada y sincronizada.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.ticker import FuncFormatter

# Configurar estilo visual
sns.set_theme(style="whitegrid")

# Convertir mes_periodo a timestamp para evitar errores en grafico de lineas
# Verificar si es necesario convertir antes de operar
if not pd.api.types.is_datetime64_any_dtype(df['mes_periodo']):
    df['mes_periodo_dt'] = df['mes_periodo'].dt.to_timestamp()
else:
    df['mes_periodo_dt'] = df['mes_periodo']

# 1. Preparar datos
# Obtener el maximo acumulado por mes y pais
df_vacunacion = df.groupby(['country_name', 'mes_periodo_dt'], observed=True)['cumulative_vaccine_doses_administered'].max().reset_index()

# ---------------------------------------------------------
# GRAFICO 1: VELOCIDAD DE DESPLIEGUE (Curva de Vacunacion)
# ---------------------------------------------------------
plt.figure(figsize=(12, 6))

sns.lineplot(
    data=df_vacunacion,
    x='mes_periodo_dt',
    y='cumulative_vaccine_doses_administered',
    hue='country_name',
    linewidth=2.5,
    palette='tab10'
)

plt.title('Avance de Vacunación: Dosis Administradas Acumuladas')
plt.xlabel('Fecha')
plt.ylabel('Dosis Totales')
plt.legend(title='País', bbox_to_anchor=(1.05, 1), loc='upper left')

# Formatear eje Y para mostrar Millones (ej: 10M, 20M)
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x/1_000_000)}M'))

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# GRAFICO 2: VOLUMEN TOTAL DE VACUNAS (Ranking Final)
# ---------------------------------------------------------
# Obtener el ultimo valor reportado (maximo global) por pais
df_total_vacunas = df.groupby('country_name', observed=True)['cumulative_vaccine_doses_administered'].max().reset_index()
df_total_vacunas = df_total_vacunas.sort_values('cumulative_vaccine_doses_administered', ascending=False)

plt.figure(figsize=(10, 6))

sns.barplot(
    data=df_total_vacunas,
    x='cumulative_vaccine_doses_administered',
    y='country_name',
    palette='viridis'
)

plt.title('Total de Dosis de Vacunas Administradas por País')
plt.xlabel('Cantidad de Dosis (Millones)')
plt.ylabel('País')

# Formatear eje X en Millones
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x/1_000_000)}M'))

# Agregar etiquetas al final de las barras
for container in plt.gca().containers:
    plt.gca().bar_label(container, fmt=lambda x: f'{int(x/1_000_000)}M', padding=3)

plt.tight_layout()
plt.show()
```
