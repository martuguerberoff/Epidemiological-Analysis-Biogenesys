```python
import pandas as pd
import os

# --- 1. CONFIGURACIÓN ---
archivo_entrada = r"C:\Users\mguer\Desktop\data_latinoamerica.csv"
archivo_salida = r"G:\Mi unidad\Soy Henry\Modulo 4\Proyecto integrador\covid_liviano.csv" # Aquí guardaremos el resultado final

location_key_paises_biogenesys = ["AR", "BR", "CL", "CO", "MX", "PE"]

# Borrar el archivo de salida si ya existe para empezar de cero
if os.path.exists(archivo_salida):
    os.remove(archivo_salida)


# --- 2. PROCESAMIENTO ---
chunk_iter = pd.read_csv(archivo_entrada, chunksize=50000, low_memory=False)

primer_bloque = True # Para saber si tenemos que escribir los encabezados (nombres de columnas)

for i, chunk in enumerate(chunk_iter):
    
    # A. Filtrar por país
    df_filtrado = chunk[chunk["location_key"].isin(location_key_paises_biogenesys)].copy()
    
    if not df_filtrado.empty:
        
        # GUARDAR EN DISCO INMEDIATAMENTE
        df_filtrado.to_csv(archivo_salida, mode='a', index=False, header=primer_bloque)
        
        primer_bloque = False 
        
    # Imprimir progreso cada 10 bloques
    if i % 10 == 0:
        print(f"Procesado bloque {i}...", end='\r')

print(" ¡TERMINADO!")
```
