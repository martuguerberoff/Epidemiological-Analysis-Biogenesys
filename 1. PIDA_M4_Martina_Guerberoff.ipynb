#2.2 FILTRADO DE PAÍSES:
Se observan 3 columnas con información acerca de los países:
- country_code
- country_name
- location_key

Se solicitan la cantidad y descripción de los valores únicos de dichas 3 columnas para elegir sobre cuál realizar el filtrado de datos.

```python
# valores únicos de la columna country_code
valores_unicos = df_covid['country_code'].unique()

print("Valores únicos de la columna country_code ", valores_unicos)
print("Cantidad de valores únicos de la columna country_code ", len(df_covid['country_code'].unique()))

# valores únicos de la columna country_name
valores_unicos = df_covid['country_name'].unique()

print("Valores únicos de la columna country_name ", valores_unicos)
print("Cantidad de valores únicos de la columna country_name ", len(df_covid['country_name'].unique()))

# valores únicos de la columna location_key
valores_unicos = df_covid['location_key'].unique()

print("Valores únicos de la columna location_key ", valores_unicos)
print("Cantidad de valores únicos de la columna location_key ", len(df_covid['location_key'].unique()))
```

```python
# Lista de países donde se expandirá Biogenesys según country_name
paises_biogenesys = ["Argentina", "Brazil", "Chile", "Colombia", "Mexico", "Peru"]

# Mantener solo las filas donde 'country_name' esté en la lista
df_limpio = df_limpio[df_limpio["country_name"].isin(paises_biogenesys)]

print("Países filtrados: Ahora el data set df_limpio contiene solo:")
print(df_limpio["country_name"].unique())

# Mostrar el total de filas y columnas
print(f"Dimensiones del dataset: {df_limpio.shape}")
```

```python
#2.3 Filtrado de datos a partir de la fecha 2021-01-01

# Convertir la columna 'date' a formato de fecha
df_limpio["date"] = pd.to_datetime(df_limpio["date"])

# Filtrado de fechas
df_limpio = df_limpio[df_limpio["date"] >= "2021-01-01"]

# Verificar los resultados
print("Filtro de fecha aplicado.")
print(f"Nuevas dimensiones del dataset: {df_limpio.shape}")
print(f"La fecha más antigua ahora es: {df_limpio['date'].min()}")
```

```python
# 1. Calculamos la cantidad de nulos por columna
nulos_cantidad = df_limpio.isnull().sum()

# 2. Calculamos el porcentaje (para entender la gravedad)
nulos_porcentaje = (df_limpio.isnull().mean() * 100).round(2)

# 3. Creamos un DataFrame para verlo bonito
df_nulos = pd.DataFrame({
    "Cantidad Nulos": nulos_cantidad,
    "Porcentaje (%)": nulos_porcentaje
})

# 4. Filtramos para ver SOLO las columnas que tienen algún nulo (ordenadas de mayor a menor)
df_nulos_filtrado = df_nulos[df_nulos["Cantidad Nulos"] > 0].sort_values("Porcentaje (%)", ascending=False)

print("--- REPORTE DE VALORES FALTANTES ---")
display(df_nulos_filtrado)
```

```python
import pandas as pd

# Ruta del archivo original
ruta_archivo = r"C:/Users/mguer/Desktop/data_latinoamerica.csv"

# Lista de países a incluir
paises_expansion = ['Argentina', 'Brasil', 'Chile', 'Colombia', 'Mexico', 'Peru']

# Nombre del archivo filtrado de salida
salida = "C:/Users/mguer/Desktop/DatosFinalesFiltrado.csv"

# Leer por partes
chunks = pd.read_csv(
    ruta_archivo,
    parse_dates=["date"],
    chunksize=1000000
)

# Filtrar y guardar por partes
for i, chunk in enumerate(chunks):
    df_filtrado = chunk[
        (chunk['country_name'].isin(paises_expansion)) &
        (chunk['date'] > '2021-01-01')
    ]

    # Guardar en el CSV final (agrega en modo 'append')
    df_filtrado.to_csv(salida, mode='a', index=False, header=(i == 0))

    print(f"Chunk {i+1} procesado: {len(df_filtrado)} filas guardadas.")

print("✅ Proceso finalizado. Archivo filtrado guardado en:")
print(salida)
```
